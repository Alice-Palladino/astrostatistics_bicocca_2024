```python
import numpy as np
import matplotlib.pyplot as plt
import math as m

deaths = np.arange(5)
groups = np.array([109, 65, 22, 3, 1])

plt.plot(deaths, groups, lw= 0.8, c = 'orange', ms = 6, marker = 'o', label = 'Data')
plt.xticks(deaths)
plt.ylabel('Groups')
plt.xlabel('Number of deaths')
plt.title('Number of deaths per group')
plt.grid(c='grey', alpha = 0.3)
plt.legend(loc='best')
plt.show()
```

```python
# Try to fit the distribution by hand (careful about the normalization!)

# Defining the Poissonian distribution:
def Poiss(mu, groups, deaths):
    P = np.empty(len(deaths))     
    for i, k in enumerate(deaths):
        P[i] = (mu**k * m.exp(-mu)) / m.factorial(k)     # Poissonian probability of having k deaths
    return P

# compute mu as sum of deaths x groups (dxg), divided the total number of groups (tg)
tg = 0
dxg = 0
for g, d in zip(groups, deaths):
    tg += g
    dxg += d*g
mu = dxg/tg
    
print('Mean: ' + str(mu))

exp_groups = Poiss(mu, groups, deaths)     # Expected number of groups (normalized) per death number (= Poissonian probability)

# Normalizing the number of groups
norm_groups = np.empty(len(groups))
for i in np.arange(len(groups)):
    norm_groups[i] = groups[i]/np.sum(groups)

plt.plot(deaths, norm_groups, ls= '', c = 'orange', ms = 6, marker = 'o', label = 'Data')
plt.plot(deaths, exp_groups, lw= 0.8, c = 'teal', ms = 3, marker = 's', label = 'Poisson')
plt.xticks(deaths)
plt.ylabel('Groups (normalized)')
plt.xlabel('Number of deaths')
plt.title('Comparing the data with the Poissonian distribution')
plt.grid(c='grey', alpha = 0.3)
plt.legend(loc='best')
plt.show()


```
